# Probability for Machine Learning

## Overview
Probability helps us model uncertainty in machine learning.
This notebook covers the fundamental probability concepts required to understand
ML algorithms such as Naive Bayes, Logistic Regression, and probabilistic models.


```python
import numpy as np
```

## Basic Probability
Probability of an event is defined as:

P(Event) = (Number of favorable outcomes) / (Total outcomes)

Example: Probability of getting an even number on a dice.


```python
# Dice example
outcomes = [1, 2, 3, 4, 5, 6]
even_numbers = [2, 4, 6]

prob_even = len(even_numbers) / len(outcomes)
print("Probabolity of getting an even number:", prob_even)
```

## Conditional Probability
Conditional probability measures the probability of an event
given that another event has already occurred.

Formula:
P(A | B) = P(A and B) / P(B)


```python
# Example dataset
students = ["Pass", "Fail", "Pass", "Pass", "Fail", "Pass"]
attendance = ["High", "Low", "High", "High", "Low", "High"]
# P(Pass | High attendance)
total_high = attendance.count("High")
pass_and_high = sum(
    1 for i in range(len(students))
    if students[i] == "Pass" and attendance[i] == "High"
)

conditional_prob = pass_and_high / total_high
print("p(Pass | High Attendance):", conditional_prob)
```

## Bayes Theorem
Bayes Theorem helps update probabilities based on new evidence.

Formula:
P(A | B) = (P(B | A) Ã— P(A)) / P(B)

This theorem is the foundation of Naive Bayes classifiers.


```python
# Given probabilities
P_A = 0.3       # Probability of disease
P_B_given_A = 0.9  # Probability of positive test if disease exists
P_B = 0.2       # Probability of positive test

# Bayes theorem
P_A_given_B = (P_B_given_A * P_A) / P_B
print("P(Disease | Positive Test):", P_A_given_B)
```

## Random Variables
A random variable assigns numerical values to outcomes of random events.
They are used to model real-world data in ML.


```python
# Random variable using numpy
samples = np.random.randint(1, 7, size = 10)
print("Random samples from dice:", samples)
```

## Key Takeaways
- Probability quantifies uncertainty
- Conditional probability is crucial for predictions
- Bayes theorem is widely used in ML models
- Random variables help represent real-world data mathematically
