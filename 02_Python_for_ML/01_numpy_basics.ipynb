# NumPy Basics for Machine Learning

## Overview
NumPy is the core Python library for numerical computing.
It is heavily used in Machine Learning for handling arrays, matrices,
and mathematical operations efficiently.


```python
import numpy as np
```

## Why NumPy?
- Faster than Python lists
- Supports vectorized operations
- Backbone of Pandas, Scikit-learn, TensorFlow


```python
# Creating NumPy arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print("1D array:", arr1)
print("2D array: \n", arr2)
```

## Array Properties
- Shape
- Size
- Data type


```python
print("Shape:", arr2.shape)
print("Size:", arr2.size)
print("Data type:", arr2.dtype)
```

## Basic Operations
NumPy supports element-wise operations.


```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

## Statistical Functions
NumPy provides fast statistical operations.


```python
data = np.array([10, 20, 30, 40, 50])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Variance:", np.var(data))
```

## Reshaping Arrays
Reshaping is common when preparing data for ML models.


```python
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape(2, 3)
print("Reshaped Array:\n", reshaped)
```

## Key Takeaways
- NumPy arrays are faster than Python lists
- Supports vectorized mathematical operations
- Essential for data preprocessing in ML
