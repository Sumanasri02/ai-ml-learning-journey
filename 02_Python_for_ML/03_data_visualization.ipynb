# Data Visualization for Machine Learning

## Overview
Data visualization helps understand patterns, trends, and relationships in data.
In Machine Learning, visualizations are used during Exploratory Data Analysis (EDA)
to make informed decisions before training models.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Why Data Visualization?
- Identify trends and patterns
- Detect outliers
- Understand relationships between features


```python
data = {
    "Age": [22, 25, 23, 24, 26, 27],
    "Score": [85, 90, 78, 88, 92, 95]
}

df = pd.DataFrame(data)
df
```

## Line Plot
Used to visualize trends over time or sequence.


```python
plt.plot(df["Age"], df["Score"])
plt.xlabel("Age")
plt.ylabel("Score")
plt.title("Age vs Score")
plt.show()
```

## Bar Plot
Used to compare categorical values.


```python
plt.bar(df["Age"], df["Score"])
plt.xlabel("Age")
plt.ylabel("Score")
plt.title("Score by Age")
plt.show()
```

## Histogram
Shows the distribution of a variable.


```python
plt.hist(df["Score"], bins=5)
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.title("Score Distribution")
plt.show()
```

## Scatter Plot
Shows the relationship between two variables.


```python
plt.scatter(df["Age"], df["Score"])
plt.xlabel("Age")
plt.ylabel("Score")
plt.title("Age vs Score Scatter")
plt.show()
```

## Seaborn Visualization
Seaborn provides high-level statistical plots.


```python
sns.scatterplot(x="Age", y = "Score", data = df)
plt.title("Seaborn Scatter Plot")
plt.show()
```

## Key Takeaways
- Visualization helps in EDA
- Detects patterns and outliers
- Essential step before ML modeling
