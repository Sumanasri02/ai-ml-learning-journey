# Linear algebra for Machine Learning

##overview
Linear Algebra is the backbone of mant ML algorithms.
this notebook covers vectors, matrices, and basic operations in python using Numpy.


```python
import numpy as np
```

## Vectors
A vector is an array of numbers representing magnitude and direction.

```python
# Example vector
v = np.array([2, 3, 4])
u = np.array([1, 0, 2])

print("vector v:", v)
print("v + u=", v + u)
print("Dot Product v.u =", np.dot(v, u))
```

## Matrices
A matrix is a 2D array of numbers used to represent data or transformations.

```python
# Example matrices
A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("A + B:\n", A + B)
print("A dot B:\n", np.dot(A, B))
print("Transpose of A:\n", A.T)
```

## Key Takeaways
- Vectors represent points or directions
- Dot product measures similarity between vectors
- Matrices store tabular data or transformations
- Matrix multiplication is fundamental in ML computations.
