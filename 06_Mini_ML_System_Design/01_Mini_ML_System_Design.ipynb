# Mini ML System Design

## Problem Statement
Build a complete ML pipeline to predict employee attrition
using a sample HR dataset.

### Workflow
1. Data Collection
2. Data Cleaning
3. Feature Engineering
4. Train-Test Split
5. Model Training
6. Evaluation
7. Deployment Idea


```python
# 1. Imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 2. Sample dataset
data = {
    "years_at_company": [1, 2, 3, 4, 5, 6, 7, 8],
    "job_satisfaction": [3, 2, 4, 3, 5, 4, 3, 5],
    "attrition": [0, 0, 0, 1, 1, 0, 1, 1]
}

df = pd.DataFrame(data)

# 3. Features and target
X = df[["years_at_company", "job_satisfaction"]]
y = df["attrition"]

# 4. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# 5. Train Model
model = LogisticRegression()
model.fit(X_train, y_train)

# 6. Predictions and Evaluation
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

## Key Takeaways
- Shows full ML workflow
- Safe train-test split
- Model evaluation
- Ready to scale with real dataset
