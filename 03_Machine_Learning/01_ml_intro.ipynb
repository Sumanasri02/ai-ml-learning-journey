# Introduction to Machine Learning

## Overview
Machine Learning (ML) enables systems to learn from data and improve
performance without being explicitly programmed.
This notebook introduces core ML concepts, types of ML,
and the overall ML workflow.


## What is Machine Learning?
Machine Learning is a subset of Artificial Intelligence that allows
computers to learn patterns from data and make predictions or decisions.

Example applications:
- Spam detection
- Recommendation systems
- Image recognition
- Stock price prediction


## Types of Machine Learning

1. **Supervised Learning**
   - Labeled data
   - Example: Regression, Classification

2. **Unsupervised Learning**
   - No labels
   - Example: Clustering, Dimensionality Reduction

3. **Reinforcement Learning**
   - Learning via rewards and penalties


## Machine Learning Workflow

1. Data collection
2. Data preprocessing
3. Feature engineering
4. Model selection
5. Training
6. Evaluation
7. Deployment


```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

#train Model
model = LinearRegression()
model.fit(X, y)

#Prediction
prediction = model.predict([[6]])
print("prediction for input 6:", prediction[0])
```

## Key Takeaways
- ML learns patterns from data
- Different types of ML solve different problems
- A structured workflow is critical for success
